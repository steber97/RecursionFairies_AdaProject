```python
import pandas as pd
import numpy as np
```

```python
#load the inputs (from angelo notebook)
inputs = pd.read_pickle('Pickles/normalDataFromAngelo.pickle')
inputs
```

```python
#load the outputs (from https://globaldatalab.org/shdi/shdi/USA/?interpolation=0&extrapolation=0&nearest_real=0&years=2017)
outputs = pd.read_csv('Data/GDL-Sub-national-HDI-data.csv')[['Region', '2017']].rename(columns={'Region': 'State', '2017':'HDI'})
outputs.State = outputs.State.apply(lambda x: x.lower())

outputs
```
